```python
import numpy as np, matplotlib.pyplot as plt
```

```python
# Length - mm
length1 = [818, 821, 820]
length2 = [602, 598, 603]
length3 = [542, 544, 545]
length4 = [878, 876, 879]
length5 = [988, 983, 989]

Lmean_mm = [np.mean(length1), np.mean(length2), np.mean(length3), np.mean(length4), np.mean(length5)]
Lmean_m = [np.mean(length1) / 1000, np.mean(length2) / 1000, np.mean(length3) / 1000, np.mean(length4) / 1000, np.mean(length5) / 1000]
Lstdev = [np.std(length1), np.std(length2), np.std(length3), np.std(length4), np.std(length5)]
Lstdev_sample = [1.5275, 2.1602, 1.5275, 1.5275, 3.2146]
Lerror = [0.7201, 1.2472, 0.7201, 0.7201, 1.5154]
Lerror_relative = [Lerror[0] / Lmean_mm[0], Lerror[1] / Lmean_mm[1], Lerror[2] / Lmean_mm[2], Lerror[3] / Lmean_mm[3], Lerror[4] / Lmean_mm[4]]
lnL_mm = [np.log(Lmean_mm[0]), np.log(Lmean_mm[1]), np.log(Lmean_mm[2]), np.log(Lmean_mm[3]), np.log(Lmean_mm[4])]
lnL_m = [np.log(Lmean_m[0]), np.log(Lmean_m[1]), np.log(Lmean_m[2]), np.log(Lmean_m[3]), np.log(Lmean_m[4])]
Lmean_mm_half = [np.sqrt(Lmean_mm[0]), np.sqrt(Lmean_mm[1]), np.sqrt(Lmean_mm[2]), np.sqrt(Lmean_mm[3]), np.sqrt(Lmean_mm[4])]
Lmean_m_half = [np.sqrt(Lmean_m[0]), np.sqrt(Lmean_m[1]), np.sqrt(Lmean_m[2]), np.sqrt(Lmean_m[3]), np.sqrt(Lmean_m[4])]

# Time - mm
period1 = [36.23, 36.47, 36.41]
period2 = [31.11, 31.33, 31.37]
period3 = [29.48, 29.77, 29.82]
period4 = [37.85, 37.73, 37.63]
period5 = [39.83, 40.02, 39.99]

Tmean_full = [np.mean(period1), np.mean(period2), np.mean(period3), np.mean(period4), np.mean(period5)]
Tmean_single = [Tmean_full[0] / 20, Tmean_full[1] / 20, Tmean_full[2] / 20, Tmean_full[3] / 20, Tmean_full[4] / 20]
Tsquare_full = [Tmean_full[0] ** 2, Tmean_full[1] ** 2, Tmean_full[2] ** 2, Tmean_full[3] ** 2, Tmean_full[4] ** 2]
Tsquare_single = [Tmean_single[0] ** 2, Tmean_single[1] ** 2, Tmean_single[2] ** 2, Tmean_single[3] ** 2, Tmean_single[4] ** 2]
Terror = [0.1020, 0.1143, 0.150, 0.0899, 0.0834]
Terror_relative = [Terror[0] / Tmean_full[0], Terror[1] / Tmean_full[1], Terror[2] / Tmean_full[2], Terror[3] / Tmean_full[3], Terror[4] / Tmean_full[4]]
lnT_full = [np.log(Tmean_full[0]), np.log(Tmean_full[1]), np.log(Tmean_full[2]), np.log(Tmean_full[3]), np.log(Tmean_full[4])]
lnT_single = [np.log(Tmean_single[0]), np.log(Tmean_single[1]), np.log(Tmean_single[2]), np.log(Tmean_single[3]), np.log(Tmean_single[4])]

def gravity(L, T):
    return np.round((4 * L * np.pi ** 2) / T ** 2, 4)

grav = [gravity(Lmean_m[0], Tmean_single[0]), gravity(Lmean_m[1], Tmean_single[1]), gravity(Lmean_m[2], Tmean_single[2]), gravity(Lmean_m[3], Tmean_single[3]), gravity(Lmean_m[4], Tmean_single[4])]
```

```python
# Plot #1: Period vs Length

print('Period mean (s)')
print('T1: ', np.round(Tmean_single[0], 4))
print('T2: ', np.round(Tmean_single[1], 4))
print('T3: ', np.round(Tmean_single[2], 4))
print('T4: ', np.round(Tmean_single[3], 4))
print('T5: ', np.round(Tmean_single[4], 4), '\n')
print('Length mean (m):')
print('L1: ', np.round(Lmean_m[0], 4))
print('L2: ', np.round(Lmean_m[1], 4))
print('L3: ', np.round(Lmean_m[2], 4))
print('L4: ', np.round(Lmean_m[3], 4))
print('L5: ', np.round(Lmean_m[4], 4), '\n')
print('Error (T[+/-] L[+/-])')
print(f'P1: Terror[{np.round(Terror[0], 4)}]  Lerror[{np.round(Lerror[0], 4)}]')
print(f'P2: Terror[{np.round(Terror[1], 4)}] Lerror[{np.round(Lerror[1], 4)}]')
print(f'P3: Terror[{np.round(Terror[2], 4)}]   Lerror[{np.round(Lerror[2], 4)}]')
print(f'P4: Terror[{np.round(Terror[3], 4)}] Lerror[{np.round(Lerror[3], 4)}]')
print(f'P5: Terror[{np.round(Terror[4], 4)}] Lerror[{np.round(Lerror[4], 4)}] \n')

x = np.polyfit(Tmean_single, Lmean_m, 1)
trend = np.poly1d(x)
print(f'Trend line slope: {np.round(x[0], 4)}')
print(f'Trend line y-int: {np.round(x[1], 4)}')

plt.figure(figsize = (20, 12))
plt.errorbar(Tmean_single, Lmean_m, xerr = Terror, yerr = Lerror, fmt = 'none', ecolor = 'blue', elinewidth = 3)
plt.plot(Tmean_single, trend(Tmean_single), color = 'red', linewidth = 3)

plt.title('Period vs Length', fontsize = 20, loc = 'left')
plt.xlabel('Period (s)', fontsize = 20)
plt.ylabel('Length (m)', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.annotate(f'Trend line: {np.round(x[0], 8)}x, y-intercept = {np.round(x[1], 4)}', (1.4, 2.0), fontsize = 25, color = 'red')
plt.grid()

#plt.savefig('PHYS 3125 Lab 2 T vs L')
```
